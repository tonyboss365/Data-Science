```python
import pandas as pd
from sklearn.datasets import load_iris
# Load iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
# Map species numbers to names
df['species'] = df['species'].map({0:'setosa', 1:'versicolor', 2:'virginica'})
# Inspect
print(df.head())
print(df.info()) 
print("col names", df.columns)
```

```python
# Numerical summary
print(df.describe())

# Categorical summary
print(df.describe(include=['object']))

# Frequency counts for a categorical column
print(df['species'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#linechart 
plt.plot(df['petal length (cm)'])
plt.xlabel("random value")
plt.ylabel("Plength")
plt.title("line chart with petal length (cm) only")
plt.show()
```

```python
#linechart 
plt.plot(df['sepal length (cm)'], df['petal length (cm)'])
plt.xlabel("Slength")
plt.ylabel("Plength")
plt.title("line chart with petal length (cm) vs sepal length (cm)")
plt.show()
```

```python
# Histogram: distribution of values
df['sepal length (cm)'].hist()
plt.xlabel("bins")
plt.ylabel("no.of elements in each bin")
plt.title("histogram of sepal length (cm)")
plt.show()
```

```python
# Plot histogram with the density curve
sns.histplot(df['sepal length (cm)'], kde=True, bins=10, color="skyblue", stat="density")

# Add labels and show
plt.title("Histogram with KDE Curve (Seaborn)")
plt.xlabel("Values")
plt.ylabel("Density")
plt.show()
```

```python
# Boxplot: detect outliers
sns.boxplot(x=df['sepal length (cm)'])
plt.show()
```

```python
#species boxplots
sns.boxplot(x="species", y="petal length (cm)", data=df)
plt.title("Petal Length Distribution by Species")
plt.show()
```

```python
# Set the style
sns.set_theme(style="whitegrid")

# Create a bar plot showing the mean sepal length for each species
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="species", y="sepal length (cm)", hue="species", palette="muted")

# Add descriptive labels
plt.title("Average Sepal Length by Iris Species")
plt.xlabel("Species")
plt.ylabel("Mean Sepal Length (cm)")

# Display the plot
plt.show()
```

```python
sns.scatterplot(data=df, x="sepal length (cm)", y="petal length (cm)", hue="species", palette="deep")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
# Create a bar plot for the frequency of the three species
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="species", hue="species", palette="Set2")

# Add descriptive labels
plt.title("Frequency Distribution of Iris Species")
plt.xlabel("Species")
plt.ylabel("Count")

# Display the plot
plt.show()
```

```python
sns.violinplot(x="species", y="petal length (cm)", data=df, palette="muted")
plt.title("Petal Width Density Profile by Species")
plt.show()
```

```python
# Correlation heatmap: relationships between numerical variables
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.show()
```

```python
# scatter plot matrix with pairplots
sns.pairplot(df, hue="species")
plt.show()
```

```python
#joint plots
sns.jointplot(data=df, x="petal length (cm)", y="petal width (cm)", hue="species", kind="scatter")
plt.show()
```
